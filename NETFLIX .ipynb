```python
## Encoding 
for i in cat_final:
    print(df[i].unique())
    print()
    print('*****')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
L= LabelEncoder()
```

```python
df['Subscription Type']=L.fit_transform(df['Subscription Type'])
```

```python
df['Country']=L.fit_transform(df['Country'])
```

```python
df['Gender']=L.fit_transform(df['Gender'])
```

```python
df['Device']=L.fit_transform(df['Device'])
```

```python
## Train and Test 

x = df.drop('Monthly Revenue',axis = 1)
y= df[['Monthly Revenue']]

xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=100)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

 ## Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
S=StandardScaler()
```

```python
xtrain_s = S.fit_transform(xtrain)
xtest_s = S.transform(xtest)
```

##  Linear Regression

```python
# Base model 
L= LinearRegression()
LR = L.fit(xtrain_s,ytrain)
```

```python
y_pred= LR.predict(xtest_s)
```

```python
LR_r2=r2_score(ytest,y_pred)
LR_r2
```

```python
LR_mse= mean_squared_error(ytest,y_pred)
LR_mse
```

## Lasso

```python
la= Lasso()
model_2 = la.fit(xtrain_s,ytrain)
```

```python
la_pred= model_2.predict(xtest_s)
```

```python
la_r2=r2_score(ytest,la_pred)
la_r2
```

```python
la_mse=mean_squared_error(ytest,la_pred)
la_mse
```

## Ridge

```python
# model 3 - Ridge

r= Ridge()
model_3 = r.fit(xtrain_s,ytrain)
```

```python
r_pred= model_3.predict(xtest_s)
```

```python
r_r2=r2_score(ytest,r_pred)
r_r2
```

```python
r_mse=mean_squared_error(ytest,r_pred)
r_mse
```

## SGD Regressor 

```python
S= SGDRegressor()
model_4 = S.fit(xtrain_s,ytrain)
```

```python
S_pred= model_4.predict(xtest_s)
```

```python
S_r2=r2_score(ytest,S_pred)
S_r2
```

```python
S_mse=mean_squared_error(ytest,S_pred)
S_mse
```

## Decision Tree Regressor

```python
D= DecisionTreeRegressor()
model_5 = D.fit(xtrain_s,ytrain)
```

```python
D_pred= model_5.predict(xtest_s)
```

```python
D_r2=r2_score(ytest,D_pred)
D_r2
```

```python
D_mse=mean_squared_error(ytest,D_pred)
D_mse
```

```python
data = {'Models':['Linear Regression','Lasso','Ridge','SGD Regressor','Decision Tree Regressor'],
                  
        'r2_score':[LR_r2,la_r2,r_r2,S_r2,D_r2],

         'MSE':[LR_mse,la_mse,r_mse,S_mse,D_mse]}

result = pd.DataFrame(data).sort_values(by =['r2_score','MSE'],ascending = False)
result
```

Conclusion:

 As the dataset is very less , the base model provides highest r2 score and minimum mean squared error. Will proceed with Linear Regression.
